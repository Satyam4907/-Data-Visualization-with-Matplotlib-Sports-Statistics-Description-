```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/sports_stats.csv")
print(df.head(10))
```

```python
df['Performance'] = df['Goals'] + df['Assists']
df['PointsPerMatch'] = df['Performance'] / df['MatchesPlayed']
```

```python
# Bar Chart: Goals per player
plt.figure(figsize=(10, 6))
plt.bar(df['Name'], df['Goals'], color='skyblue')
plt.title('Goals per Player')
plt.ylabel('Goals')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Line Chart: Points per match
plt.figure(figsize=(10, 6))
plt.plot(df['Name'], df['PointsPerMatch'], marker='o', color='green')
plt.title('Points Scored per Match')
plt.ylabel('Points per Match')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Pie Chart: Player position distribution
position_counts = df['Position'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Player Position Distribution')
plt.axis('equal')
plt.show()
```

```python
# Scatter Plot: Age vs Performance
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Performance'], color='red', s=100)
for i in range(len(df)):
    plt.text(df['Age'][i], df['Performance'][i], df['Name'][i], fontsize=9)
plt.title('Age vs Performance')
plt.xlabel('Age')
plt.ylabel('Performance (Goals + Assists)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
